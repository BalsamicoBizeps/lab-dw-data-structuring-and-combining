```python
import pandas as pd

# Load the dataset
url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
df_1 = pd.read_csv(url1)

url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
df_2 = pd.read_csv(url2)

df_1_clean = pd.read_csv("output.csv")
```

```python
df_1_clean = df_1_clean.drop(columns="Unnamed: 0", axis = 1)
df_1_clean.head()
```

```python
df_2.columns = pd.Series(df_2.columns).apply(lambda col: col.replace(" ", "_").lower())
df_2.rename(columns= {'st': 'state'}, inplace=True)
df_2.head()

df_1.columns = pd.Series(df_1.columns).apply(lambda col: col.replace(" ", "_").lower())
df_1.rename(columns= {'st': 'state'}, inplace=True)
```

```python
df_1["vehicle_class"].value_counts()
```

```python
df_1=df_1.dropna(how='all')
```

```python


df_1["gender"] = df_1["gender"].replace(["F", "Femal", "M", "Male"],[ "female", "female", "male", "male"])

df_1["state"] = df_1["state"].replace("AZ", "Arizona")
df_1["education"] = df_1["education"].replace("Bachelors", "Bachelor")
df_1["customer_lifetime_value"] = df_1["customer_lifetime_value"].apply(lambda x: str(x).rstrip("%"))

df_1["total_claim_amount"] = df_1["total_claim_amount"].apply(lambda x: int(x)if type(x) != str else x)
df_1["vehicle_class"] = df_1["vehicle_class"].replace(["Sports Car", "Luxury SUV", "Luxury Car"],[ "Luxury", "Luxury", "Luxury"])

df_1["number_of_open_complaints"] = df_1["number_of_open_complaints"].str.split("/").str[1]
df_1['number_of_open_complaints'] = pd.to_numeric(df_1['number_of_open_complaints'], errors='coerce')
df_1["number_of_open_complaints"] = df_1["number_of_open_complaints"].apply(lambda x: int(x)if type(x) != str else x)
```

```python
df_1["gender"] = df_1["gender"].fillna("unknown")
```

```python
df_1.isna().any()
```

```python
df_1_clean["customer_lifetime_value"] = df_1_clean["customer_lifetime_value"].apply(lambda x: int(x)if type(x) != str else x)
df_1_clean["income"] = df_1_clean["income"].apply(lambda x: int(x)if type(x) != str else x)
df_1_clean["monthly_premium_auto"] = df_1_clean["monthly_premium_auto"].apply(lambda x: int(x)if type(x) != str else x)
df_1_clean["number_of_open_complaints"] = df_1_clean["number_of_open_complaints"].apply(lambda x: int(x)if type(x) != str else x)
df_1_clean["total_claim_amount"] = df_1_clean["total_claim_amount"].apply(lambda x: int(x)if type(x) != str else x)
```

```python
df_merged = pd.concat([df_1_clean, df_2], axis = 0)
df_merged
```

```python
df_merged["policy_type"].value_counts()
```

```python
df_2["customer_lifetime_value"] = df_2["customer_lifetime_value"].round(2).astype(int)
df_2["total_claim_amount"] = df_2["total_claim_amount"].round(2).astype(int)
```

```python
url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
df_marketing_customer = pd.read_csv(url3)
df_marketing_customer.head()
```

```python
df_marketing_customer["monthly_premium_auto"].value_counts()
```

```python
df_marketing_customer["customer_lifetime_value"] = df_marketing_customer["customer_lifetime_value"].round(2)
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
pivot_df = df_marketing_customer.pivot_table(columns='sales_channel', values=['customer_lifetime_value'], aggfunc='sum')
pivot_df
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
pivot_avg = df_marketing_customer.pivot_table(index = 'education', columns='gender', values=['customer_lifetime_value'], aggfunc='mean')
pivot_avg
```
